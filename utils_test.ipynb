```python
%matplotlib inline
import os
import matplotlib
import pylab
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd

wav, sr = librosa.load('dataset/train/LJ001-0002.wav')   
# make pictures name 
save_path = 'test.jpg'
print(sr)
print(wav.shape)
plt.figure(figsize=(14, 5))
librosa.display.waveplot(wav, sr=sr)
```

```python
wav, sr = librosa.load('dataset/train/LJ001-0004.wav') 
X = librosa.stft(wav)
Xdb = librosa.amplitude_to_db(abs(X))
print(X.shape)
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```
